## Investigation Overview
>In this investigation I will provide visualization between some variables to see their influence on other variables such as the trip duration


## Dataset Overview

> This data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area.



```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_clean = pd.read_csv('df_clean.csv')
```

## Gender for each trip 

> As we can see males use bike more than females

```python
sb.countplot(data=df_clean, x='member_gender', color=sb.color_palette()[0]);
plt.xlabel('Gender');
plt.ylabel('Nmber of trips');
plt.title('Number of trips for gender ');
```

## User type for each trip 

> This show subscribers users more usage bike

```python
plt.hist(data=df_clean, x='user_type', color=sb.color_palette()[0]);
plt.title('Number of trips for user type  ');
plt.xlabel('user type ');
plt.ylabel('Nmber of trips');
```

## Duration trip according to weekdays

When looking at each day we can see the longest duration trip on saturday and sunday weekend

```python
sb.pointplot(data = df_clean , x='day',y='duration_min',order =["Mon", "Tue", "Wed", "Thu", "Fri", "Sat","Sun"],color=sb.color_palette()[0]);
plt.title('duration of trip according to weekdays');
```

## Trip duration across user type in weekdays

looking at user type customers have longer duration trip comparing to subscribers in each days

```python
fig = plt.figure(figsize = [7,4]);
sb.pointplot(data = df_clean, x = 'day', y = 'duration_min',order =["Mon", "Tue", "Wed", "Thu", "Fri", "Sat","Sun"], hue = 'user_type');
plt.title('Trip duration across user type in  different days ');
```

## Trip duration across gender type in different days
looking at gender type female have longer duration trip than male in all days 


```python
fig = plt.figure(figsize = [8,6]);
sb.pointplot(data = df_clean, x = 'day', y = 'duration_min',order =["Mon", "Tue", "Wed", "Thu", "Fri", "Sat","Sun"], hue = 'member_gender',  ci=None);
plt.title('Trip duration across gender type in different days');
```
